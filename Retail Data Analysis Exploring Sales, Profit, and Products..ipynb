### what is the proportion of total sales (%) in South , West , Central , East in 2018 - 2020

```python
regions = ['South', 'West', 'Central', 'East']
years = [2018, 2019, 2020]

for year in years:
    # Filter data for the specific year
    filtered_df = df[df['Order Date'].dt.year == year]

    # Calculate total sales for all regions in the year
    total_sales_year = filtered_df['Sales'].sum()

    print(f"Year: {year} ", end="")  # Print year at the beginning of the row

    for region in regions:
        # Calculate total sales for the region in the year
        region_sales_year = filtered_df[filtered_df['Region'] == region]['Sales'].sum()

        # Calculate proportion of total sales for the region in the year
        proportion_region_year = (region_sales_year / total_sales_year) * 100

        print(f"{region}: {proportion_region_year:.2f}% ", end="")  # Print region and proportion

    print()  # Move to the next line for the next year
```

```python
import matplotlib.pyplot as plt

years = [2018, 2019, 2020]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns, adjust figsize as needed

# Iterate through years and create pie charts
for i, year in enumerate(years):
    # Filter data for the specific year
    filtered_df = df[df['Order Date'].dt.year == year]

    # Calculate total sales by region for the year
    sales_by_region = filtered_df.groupby('Region')['Sales'].sum()

    # Create the pie chart within the subplot
    axes[i].pie(sales_by_region.values, labels=sales_by_region.index, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Proportion of Total Sales by Region in {year}')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by year and region
sales_by_year_region = df.groupby([df['Order Date'].dt.year, 'Region'])['Sales'].sum().reset_index()

# Pivot the data
sales_pivot = sales_by_year_region.pivot(index='Region', columns='Order Date', values='Sales')

# Create the line chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
for region in sales_pivot.index:
    plt.plot(sales_pivot.columns, sales_pivot.loc[region], label=region)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Region over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Filter data for the period 2018-2020
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]

# Group data by region and year and calculate total sales
sales_by_region_year = filtered_df.groupby(['Region', filtered_df['Order Date'].dt.year])['Sales'].sum().reset_index()

# Get unique regions and years
regions = sales_by_region_year['Region'].unique()
years = sales_by_region_year['Order Date'].unique()

# Set bar width and positions
bar_width = 0.2
index = np.arange(len(regions))

# Create the bar chart
plt.figure(figsize=(10, 6))  # Adjust figure size if needed

# Plot bars for each year
for i, year in enumerate(years):
    sales_year = sales_by_region_year[sales_by_region_year['Order Date'] == year]['Sales']
    plt.bar(index + i * bar_width, sales_year, bar_width, label=str(year))

# Set labels, title, and legend
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Total Sales by Region and Year (2018-2020)')
plt.xticks(index + bar_width, regions)
plt.legend()

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# Plot 2: Sales Trend over Time (2018-2020)
df['Order Month'] = df['Order Date'].dt.to_period('M')
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]
sales_by_month = filtered_df.groupby('Order Month')['Sales'].sum()
plt.figure(figsize=(10, 6))
plt.plot(sales_by_month.index.astype(str), sales_by_month.values)
plt.xlabel("Order Month")
plt.ylabel("Total Sales")
plt.title("Sales Trend over Time (2018-2020)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### which Segment has the highest profit in 2018 - 2020

```python
df["Segment"].unique()
```

```python
# which Segment has the highest profit in 2018 - 2020
# Filter data for the period 2018-2020
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]

# Group data by segment and calculate total profit
segment_profit = filtered_df.groupby('Segment')['Profit'].sum()

# Find the segment with the highest profit
highest_profit_segment = segment_profit.idxmax()

print(f"The segment with the highest profit from 2018 to 2020 is: {highest_profit_segment}")
```

```python
import matplotlib.pyplot as plt

# Filter data for the period 2018-2020
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]

# Group data by segment and calculate total profit
segment_profit = filtered_df.groupby('Segment')['Profit'].sum()

# Create the bar chart
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.bar(segment_profit.index, segment_profit.values)
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.title('Total Profit by Segment (2018-2020)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

### find top 10 popular products in terms of number of orders vs. total sales during 2018-2020

```python
# find top 10 popular products in terms of number of orders vs. total sales during 2018-2020
filtered_df = df[(df['Order Date'].dt.year >= 2018) & (df['Order Date'].dt.year <= 2020)]
product_rank = filtered_df.groupby('Product Name').agg(
    total_orders=('Order ID', 'nunique'),  # 'nunique' counts unique orders
    total_sales=('Sales', 'sum')
)
top_10_orders = product_rank.sort_values(by=['total_orders', 'total_sales'], ascending=False).head(10)
top_10_sales = product_rank.sort_values(by=['total_sales', 'total_orders'], ascending=False).head(10)
print("Top 10 products by number of orders:\n", top_10_orders)
print("\nTop 10 products by total sales:\n", top_10_sales)
```
